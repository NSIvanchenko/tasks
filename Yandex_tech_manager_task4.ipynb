```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv('data_task4_old.txt', sep="\t", header=0)
data
```

```python
data.dtypes
```

```python
data['assigned_ts'] = pd.to_datetime(data['assigned_ts'])
data['closed_ts'] = pd.to_datetime(data['closed_ts'])
data.head()
```

```python
data.dtypes
```

```python
data['duration_per_Microtask'] = (data['closed_ts'] - data['assigned_ts']) / data['Microtasks']
data['duration_per_Microtask'] = data['duration_per_Microtask'].dt.total_seconds().astype(int)

# Длительность выполнения одного микрозадания указана в секундах
# https://stackoverflow.com/questions/40992976/python-convert-datetime-column-into-seconds
```

```python
data.head()
```



**Посмотрим на распределение длительности выполнения одного микрозадания:**


```python
data.hist(column = 'duration_per_Microtask', bins=25, grid=False, figsize=(12,8), color='#86bf91', zorder=2, rwidth=0.9)
```

**Поскольку изображеие не дает в полной мере понять, как распределен показатель, то прологарифмируем его для наглядности распредеелния и построим новый график**

```python
data['log_duration'] = np.log1p(data['duration_per_Microtask'])
data.hist(column = 'log_duration', bins=25, grid=False, figsize=(12,8), color='#86bf91', zorder=2, rwidth=0.9)
plt.ylabel('Frequency')
plt.xlabel('Log duration')
plt.title('Log duration per Microtask')
```

```python
data.describe()
```

**Посмотрим на хвосты распределния (максимальные и минимальные временные промежутки выполнения одного задания)**

```python
max_duration = data.sort_values(by=['duration_per_Microtask'], ascending=False)
```

```python
max_duration.head(30)
```

###### Посмотрим на наименьшее время выполнения одного микрозадания и убедимся, что тоже есть выбросы. Скроее всего невозможно выполнить одно задание за такой короткий срок (тем более выполнить задание до его выдачи:)).

```python
min_duration = data.sort_values(by=['duration_per_Microtask'], ascending=True)
min_duration.head(30)
```

**Таким образом, необходимо исключить из рассмотрения такие аномальные показатели времени выполнения одного микрозадания.**

Для определения оптимальной ставки оплаты за одно задание первая мысль, которая приходит в голову, это взять среднее время выполнения одного задания по всем пользователям и поделить это значение на заранее заданные нам 30 секунд, оплата по которым составляет N рублей.

Среднее значение времени выполнения одного задания по всем асессорам составляет примерно 568 секунд (как видно ниже). Соответственно оплата одной задачи будет составлять 568 / 30 = 18,93 * N рублей.
